```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Let's create 2 numpy arrays

x=np.linspace(start=0,stop=10,num=5)
y=x**2
```

```python

print(x)
print(y)
```

```python
#Now if we want to plot the relationship between x and y, we can have it with simple 
plt.plot(x,y)
```

```python
# If you are using any other editor apart form jupyter, then you have add an additional command as plt.show() to show the plot
```

```python
# If you want to add x-label and y-label then you can add it using
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
# We can also add title using plt.title
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
```

```python
# We can have multiple plots on the same plane as well using subplot method

plt.subplot(1,2,1) # It takes no. of rows as the first argument, no. of columns as second and plot name as third
plt.plot(x,y)

#Let's take another one subplot
plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
# we can also provide different colors to distinguish

plt.subplot(1,2,1)
plt.plot(x,y,'r') # Here the color is red

plt.subplot(1,2,2)
plt.plot(y,x,'g') # Here the color is green
```

```python
#There is another way of ploting using matplotlib which is called as the object oriented method
```

```python
#For that we have to define a figure or a canvas you can say on which we will draw the plot

fig=plt.figure()
```

```python
#Now we can draw on the fig object
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) # Here we are providing 4 values left,bottom, width and height
```

```python
#We can now plot the same x,y using axes
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
# we can also add the x and y label and also the title
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title')
```

```python
#We can also create multiple fighures also using this object oriened approach

fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8]) # here we are taking 10% of left, 10% of bottom, 80% of width and 80% of height
axes1.set_title('axes1')

axes2=fig.add_axes([0.5,0.2,0.3,0.3]) # here we are taking 50% of left, 20% of bottom, 30% of width and 30% of height
axes2.set_title('axes2')
```

```python
# we can also plot x, y in axes1 and axes2

fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8]) 
axes1.set_title('axes1')
axes1.plot(y,x)

axes2=fig.add_axes([0.5,0.2,0.3,0.3]) 
axes2.set_title('axes2')
axes2.plot(x,y)
```

```python
# we can also create the subplots using plt.subplots()

fig,axes=plt.subplots()
```

```python
# here under the subplots() function we can add the no. of rows and columns

fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
# So here the axes does the axes creartion task that we were doing in axes=fig.add_axes([0.5,0.2,0.3,0.3])

# now let's check the axes
axes
```

```python
# So the axes is an array of list of subplots, so we can iterate through it

fig,axes=plt.subplots(nrows=1,ncols=2)
for i in axes:
    i.plot(x,y)
```

```python
# we can also iterate using index notation

fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(y,x)
axes[1].plot(x,y)
```

```python
# in case of any overlaping we can use plt.tight_layout()

fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(y,x)
axes[1].plot(x,y)
plt.tight_layout()
```

```python
# we can also specify the figure size

fig=plt.figure(figsize=(2,5))
```

```python
# now we can also add axes

fig=plt.figure(figsize=(2,5))

axes=fig.add_axes([0,0,1,1])
```

```python
# now we can also plot as we did before

fig=plt.figure(figsize=(2,5))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
# similarly we can do the same for subplots, everything is same as it was in case of subplots the only thing is figsize is a 
# new argument added

fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(7,2))
axes[0].plot(y,x)
axes[1].plot(x,y)
plt.tight_layout()
```

```python
# adding legend
```

```python
# if we have a fig

fig=plt.figure(figsize=(8,2))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2)
axes.plot(x,x**3)
```

```python
# We have 2 lines here but we don't know which one is square and which one is cubed
# to know that we need to add legend

fig=plt.figure(figsize=(8,2))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2)
axes.plot(x,x**3)
axes.legend()
```

```python
# notice that 'No handles with labels found to put in legend' is the error we are receiving
# to deal with it

fig=plt.figure(figsize=(8,2))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='square')
axes.plot(x,x**3,label='cube')
axes.legend()
```

```python
# Now we can be able to know which one is square and which one is cubed
```

```python
#color
```

```python
# we can specify the colors of the plot

fig=plt.figure(figsize=(8,2))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple')
```

```python
#We can also have RGB hex code for the color references

fig=plt.figure(figsize=(8,2))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#32a852') # and this will give us green color
```

```python
# we can alos specify the line width

fig=plt.figure(figsize=(8,2))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',linewidth=3) # the default is always 1, we can change it as per our requirements
```

```python
# we can make the little dull by using the alpha command

fig=plt.figure(figsize=(8,2))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',linewidth=3,alpha=0.5)
```

```python
# we can also have linestyle

fig=plt.figure(figsize=(8,2))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',linewidth=3,alpha=0.5,linestyle='--')
```

```python
# now if we want to get the points location on the plot then we can do that by using markers

fig=plt.figure(figsize=(8,2))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',linewidth=3,alpha=0.5,marker='o') # 'o' is the basic marker which specifies as dot points
```

```python
# we can specify the marker size to make it look bigger or smaller

fig=plt.figure(figsize=(8,2))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',linewidth=3,alpha=0.5,marker='*',markersize=10) 
```

```python
# we can customize the marker as well like adding the marker color, marker edge color and width

fig=plt.figure(figsize=(8,2))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',linewidth=1,alpha=0.5,marker='o',markersize=20,markerfacecolor='red',markeredgewidth=2,
         markeredgecolor='green') 
```
